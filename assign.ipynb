```python
import numpy as np 
import nltk
from nltk.corpus import brown
import re
from nltk.tokenize import word_tokenize
import pandas as pd
```

```python
words = brown.tagged_words(categories='lore',tagset='universal')
pos = ([i[1] for i in words])
postag=list(set(pos))
print(postag)
```

<h1> Sequence of Pos-tag

```python

def ngram(n):
    hmm_list=[]
    ml=[]
    for i in pos:
        hmm_list.append(i)
        if len(hmm_list)==n:
            hmm_list=hmm_list[1:]
        ml.append(hmm_list)
    return ml
```

<h1> Finding the pos-tag of the word in a sentence

```python
word_list=[]
for i in words:
    j=list(i)
    word_list.append(j)
```

```python
text='India is the best country in the world'
a=text.split()
```

```python
var1=[]
for i in a:
    var2=[]
    for j in postag:
        var3=[i,j]
        var2.append(var3)
    var1.append(var2)
```

```python
count11=[]
for i in var1:
    var4=i
    count1=[]
    for j in var4:
        count=word_list.count(j)
        count1.append(count)
    count11.append(count1)
```

```python
countlist=[]
for i in count11:
    count111=i
    var5=postag[count111.index(max(count111))]
    countlist.append(var5)
```

```python
dict={'word':a, 'pos':countlist}
df=pd.DataFrame(dict)
df
```
