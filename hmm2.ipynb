```julia
import numpy as np 
import nltk
from nltk.corpus import brown
import re
from nltk.tokenize import word_tokenize
import pandas as pd
```

```julia
words = brown.tagged_words(categories='lore',tagset='universal')
pos = ([i[1] for i in words])
```

```julia
def hmm(text, n_gram=1):
    markov = {}
    for i in range(len(text)-n_gram-1):
        curr_state, next_state ='',''
        for j in range(n_gram):
            curr_state+=text[i+j]+" "
            next_state+=text[i+j+n_gram]+" "
            curr_state=curr_state[:-1]
            next_state=next_state[:-1]
        if curr_state not in markov:
            markov[curr_state] = {}
            markov[curr_state][next_state] = 1
        else:
            if next_state in markov[curr_state]:
                markov[curr_state][next_state]+=1
            else:
                markov[curr_state][next_state] = 1

    for curr_state,transition in markov.items():
        total = sum(transition.values())
        for state, count in transition.items():
            markov[curr_state][state] = count/total
    return markov
markov = hmm(pos)
```

```julia
L = len(set(pos))
A = np.zeros([L,L])
for i in range(len(set(pos))):
    for j in range(len(set(pos))):
        x = list(set(pos))[i]
        y = list(set(pos))[j]
        if y in markov[x]:
            A[i][j] = markov[x][y]
        else:
            A[i][j] = 0
A_ = pd.DataFrame(A)
A_.columns = list(set(pos))
A_.index = list(set(pos))
print("The transition matrix is following:\n")
A_
```
